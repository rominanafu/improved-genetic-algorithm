```python
benchmarks = ['C101', 'C102', 'C103', 'C104', 'C105', 'C106', 'C107', 'C108', 'C109', 
              'C201', 'C202', 'C203', 'C204', 'C205', 'C206', 'C207', 'C208',
              'R101', 'R102', 'R111', 'R112']
datetimes = ['2025-10-22_16-36-44', '2025-10-22_16-42-01', '2025-10-22_16-46-20', '2025-10-22_16-50-49', '2025-10-22_16-58-45', '2025-10-22_17-03-35', '2025-10-22_17-17-50', '2025-10-22_17-23-00', '2025-10-22_17-28-12', 
             '2025-10-22_17-35-14', '2025-10-22_18-18-22', '2025-10-22_22-04-03', '2025-10-22_22-22-00', '2025-10-22_22-37-53', '2025-10-22_22-57-25', '2025-10-22_23-10-42', '2025-10-22_23-25-31',
             '2025-10-22_18-32-17', '2025-10-22_18-37-33', '2025-10-22_21-03-46', '2025-10-22_21-10-55']

solutions = [828.94] * 9
solutions += [591.56] * 2 + [591.17] + [590.60] + [588.88] + [588.49] + [588.29] + [588.32]
solutions += [1650.80] + [1486.12] + [1096.72] + [982.14]

for i in range(len(benchmarks)):
    bn = benchmarks[i]
    dt = datetimes[i]
    sol = solutions[i]
    
    # Read data
    iga = pd.read_csv(f'../results/IGA_{bn}_{dt}.csv')
    igam = pd.read_csv(f'../results/IGAm_{bn}_{dt}.csv')

    # Plot best evaluation
    n = len(iga)
    x = np.arange(n)
    plt.plot(x, iga['Best Evaluation'], color='blue', label='IGA')
    plt.plot(x, igam['Best Evaluation'], color='red', label='IGAm')
    plt.plot(x, [sol]*n, mfc='w', mec='g', marker='o', label='Optimum')
    plt.title(f'Best Evaluation, benchmark {bn}')
    plt.legend()
    plt.show()
```
