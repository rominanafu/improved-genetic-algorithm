```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Best evaluation

```python
benchmarks = ['C101', 'C102', 'C103', 'C104']
datetimes = ['2025-10-16_16-17-13', '2025-10-16_21-03-45', '2025-10-16_21-09-23', '2025-10-16_21-26-35']
solutions = [828.94] * 4

for i in range(4):
    bn = benchmarks[i]
    dt = datetimes[i]
    sol = solutions[i]
    
    # Read data
    iga = pd.read_csv(f'../results/IGA_{bn}_{dt}.csv')
    igam = pd.read_csv(f'../results/IGAm_{bn}_{dt}.csv')

    # Plot best evaluation
    n = len(iga)
    x = np.arange(n)
    plt.plot(x, iga['Best Evaluation'], color='blue', label='IGA')
    plt.plot(x, igam['Best Evaluation'], color='red', label='IGAm')
    plt.plot(x, [sol]*n, mfc='w', mec='g', marker='o', label='Optimum')
    plt.title(f'Best Evaluation, benchmark {bn}')
    plt.legend()
    plt.show()
```
